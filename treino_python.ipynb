## Realizando a Leitura da Base de dados

```python
df = pd.read_excel("treino_excel.xlsx")
df.head(5)
```

## Tratando dados 

```python
# Verificar quais colunas tem no dataset
df.columns
```

```python
df[' Envolvimento dos Parentes'].str.strip() # Para tirar espaços em branco de uma coluna específica
df.columns = df.columns.str.strip() # Para remover espaço em branco de todas as colunas
df.head(5)
```

```python
# Find and replace
df['Parental_Involvement'] = df['Parental_Involvement'].replace(['High', 'Medium', 'Low'], ['Alto', 'Médio', 'Baixo'])
df['Access_to_Resources'] = df['Access_to_Resources'].replace(['High', 'Medium', 'Low'], ['Alto', 'Médio', 'Baixo'])

# OU
df[['Parental_Involvement', 'Access_to_Resources', 'Motivation_Level']] = df[['Parental_Involvement', 'Access_to_Resources', 'Motivation_Level']].replace(
    {'High': 'Alto', 'Medium': 'Médio', 'Low': 'Baixo'}
)

df.head(5)
```

```python
# Removendo uma coluna do dataset
df = df.drop('Column1', axis="columns", errors='ignore')
df.head(5)
```

```python
# Criar uma nova coluna baseado em alguma condição:

import numpy as np

df['StatusHoras'] = np.where(df['Hours_Studied'] > 10, 'Estudou Muito', 'Estudou Pouco')
df.head(5)

'''
Obs: E se fosse preciso mais condições:

# Definir as condições
conditions = [
    (df['Score'] >= 90),
    (df['Score'] >= 70) & (df['Score'] < 90),
    (df['Score'] < 70)
]

# Definir os resultados correspondentes
choices = ['Excelente', 'Aprovado', 'Reprovado']

# Criar a nova coluna com base nas condições
df['Status'] = np.select(conditions, choices)
'''
```

```python
# Verificar quantos tem de cada categoria
df['StatusHoras'].value_counts()
```

```python
# Filtrar os dados
df_filtrado = df.loc[:,['Hours_Studied', 'Attendance', 'Envolvimento dos Parentes', 'Motivation_Level']]
df_filtrado.head()
```

```python
# Funções de agregação

# Realizado o agrupamento
analise = df_filtrado.groupby(['Envolvimento dos Parentes']).agg({'Hours_Studied' : ['mean']})

# Ordenando e arredondando
analise = analise.sort_values(('Hours_Studied', 'mean'), ascending=False).round(2)

'''
Mais de uma coluna:
analise = df_filtrado.groupby(['Envolvimento dos Parentes']).agg({'Hours_Studied' : ['mean']})
'''

analise
```

### Importando biblotecas necessárias

```python
#pip install requests beautifulsoup4
import requests
from bs4 import BeautifulSoup
```

### Estabelecendo conexão com o site

```python
# URL do site
url = "https://www.globo.com/"

# Fazer a requisição para o site
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    
    # Criar o objeto BeautifulSoup para fazer o parsing do HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Exibir o HTML para analisar a estrutura (opcional)
    #print(soup.prettify())
else:
    print(f"Erro ao acessar o site: {response.status_code}")
```

### Extraindo dados

```python
# Ver se está dando certo
h2_tag = soup.find_all('h2')
h2_tag
```

```python
# Listas para armazenar os dados
titulos = []
descricoes = []

# Extrair os títulos e as descrições
for noticia in soup.find_all('h2', class_='post__title'):
    #print(noticia)
    
    titulo = noticia.text.strip()


    if titulo:
        titulos.append(titulo)

# Exibir os resultados
print("Títulos:", titulos)
```

### Colocando em uma base de dados simples

```python
df = pd.DataFrame(titulos, columns=['Título'])
df.head(5)
```

### Agora fazendo tudo de uma vez:

```python
# Lista de URLs de notícias (substitua pelos seus links reais)
urls = [
    "https://g1.globo.com/politica/noticia/2024/10/18/stf-forma-maioria-para-proibir-revista-intima-vexatoria-em-presidios.ghtml",
    "https://ge.globo.com/ba/futebol/times/vitoria/noticia/2024/10/18/presidente-do-vitoria-detona-o-corinthians-e-cobra-punicao-no-brasil-loucura-tem-que-acabar.ghtml",
]

# Lista para armazenar os títulos e conteúdos
noticias = []

for url in urls:
    
    response = requests.get(url) 
    
    if response.status_code == 200:  
        soup = BeautifulSoup(response.content, 'html.parser')
        
        titulo = soup.find('h1', class_='content-head__title').text.strip()
        

        conteudo_paragrafos = []
        
        for p in soup.find_all('p'):  
            texto_paragrafo = p.text.strip() 
            
            if texto_paragrafo:
                conteudo_paragrafos.append(texto_paragrafo)
        
        # Junta todos os parágrafos em uma única string, separados por espaços
        conteudo = ' '.join(conteudo_paragrafos)
        
        # Adicionar os dados à lista
        noticias.append({
            'Título': titulo,
            'Conteudo': conteudo
        })
        
    else:
        print(f"Erro ao acessar {url}: {response.status_code}")
```

```python
df_noticias = pd.DataFrame(noticias)
df_noticias.head(5)
```
